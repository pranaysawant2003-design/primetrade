# Trader Performance vs Bitcoin Market Sentiment

**Objective:** Join daily Bitcoin Fear & Greed sentiment with Hyperliquid fill-level trader data, measure how performance changes across sentiment regimes, and identify actionable trading patterns.

The main metrics are closed PnL, win rate on realized fills, PnL per realized fill, PnL after fees, and PnL/volume efficiency in basis points.

## Data assumptions

- Join on calendar date using `Timestamp IST`.
- Treat the Hyperliquid file as fill-level data.
- Calculate win rate only where `Closed PnL != 0`.
- Do not analyze leverage because the uploaded trade file does not contain a leverage column.

```python
from pathlib import Path
from analysis import analyze

# Rename or replace these paths with the company-provided files.
TRADES = Path('historical_trader_data.csv')
SENTIMENT = Path('bitcoin_sentiment.csv')
OUTDIR = Path('analysis_output')

results = analyze(TRADES, SENTIMENT, OUTDIR)
results['overall']
```

```python
# Sentiment-regime comparison
for row in results['sentiment_summary']:
    print(
        row['sentiment_class'],
        'PnL=', round(row['closed_pnl'], 2),
        'Win rate=', round(row['win_rate'], 4),
        'PnL margin bps=', round(row['pnl_margin_bps'], 2),
    )
```

```python
# Daily sentiment relationships
for row in results['correlations']:
    print(row['metric'], 'rho=', round(row['rho'], 4), 'p=', round(row['p_value'], 4))
```

## Conclusions from the supplied dataset

1. Sentiment does not meaningfully explain absolute daily PnL by itself.
2. Extreme Greed has the highest PnL efficiency after normalizing for trading volume.
3. Close Short performance is much stronger in Fear than in Greed.
4. Non-crossed execution substantially outperforms crossed execution and incurs much lower fees.
5. The strongest strategy is therefore **sentiment regime + direction + execution style**, not sentiment alone.
